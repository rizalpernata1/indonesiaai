## Buatlah sebuah program yang dapat menyimpan kontak dan juga menampilkan kontak-kontak tersebut. Setiap kontak terdiri dari nama dan nomor telepon. Saat menjalankan program, program akan menampilkan pesan sambutan dan juga menampilkan menu. Program akan secara otomatis kembali menu apabila user telah selesai melihat kontak ataupun menambahkan kontak.

```python
NamaKontak = ['Andi', 'Budiman']
NoTelepon = ['0112233445', '0556677889']

def DaftarKontak():
    print('\nDaftar Kontak : ')
    for i in range(len(NamaKontak)):
        print(f'Nama: {NamaKontak[i]}')
        print(f'No Telepon : {NoTelepon[i]}')
              
def TambahKontak():
    print('\nMasukkan Data Kontak : ')
    NamaKontak.append(input('Name : '))
    NoTelepon.append(input('No Telepon : '))
    print('Kontak berhasil ditambahkan')
    
    
print('Selamat Datang!')
while True:
    print('''\n--- Menu ---
    1. Daftar Kontak
    2. Tambah Kontak
    3. Keluar\n
    ''')
    
    menu = int(input('Pilih Menu : '))
    if menu == 1:
        DaftarKontak()
    
    elif menu == 2:
        TambahKontak()
    
    elif menu == 3:
        print('\nProgram selesai, sampai jumpa!')
        break
        
    else:
        print('\nMenu tidak tersedia')
```
